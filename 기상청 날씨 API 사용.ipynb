```python
#API 호출하기
import requests
## 데이터 활용 https://www.data.go.kr/data/15059093/openapi.do
url = 'http://apis.data.go.kr/1360000/AsosDalyInfoService/getWthrDataList'
key = 'your_api_key'
params ={'serviceKey' : key, 
         'pageNo' : '1', 
         'numOfRows' : '999', 
         'dataType' : 'JSON', 
         'dataCd' : 'ASOS', 
         'dateCd' : 'DAY', 
         'startDt' : '{}0101'.format(2000), 
         'endDt' : '{}1230'.format(2010),  
         'stnIds' : '138' }

response = requests.get(url, params=params)
result = json.loads(response.text)
df=json_normalize(result['response']['body']['items']['item'])
df = df[['tm','stnNm','sumRn','maxWd']]
```

```python
#API 호출하기
import requests
from tqdm import tqdm
year =range(2001,2022)

for i in tqdm(year):
    url = 'http://apis.data.go.kr/1360000/AsosDalyInfoService/getWthrDataList'
    key = 'your_api_key'
    params ={'serviceKey' : key, 
             'pageNo' : '1', 
             'numOfRows' : '999', 
             'dataType' : 'JSON', 
             'dataCd' : 'ASOS', 
             'dateCd' : 'DAY', 
             'startDt' : '{}0101'.format(i), 
             'endDt' : '{}1230'.format(i),  
             'stnIds' : '138' }

    response = requests.get(url, params=params)
    result = json.loads(response.text)
    dff=json_normalize(result['response']['body']['items']['item'])
    dff = dff[['tm','stnNm','sumRn','maxWd']]
    df = pd.concat([df,dff])
```
